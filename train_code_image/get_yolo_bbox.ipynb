```python
import os
import pandas as pd
```

```python
df = pd.read_csv('./data-640/train_image_level.csv')
df
```

```python
df_meta = pd.read_csv('./data-640/meta.csv')
df_meta = df_meta[df_meta.split == 'train'].reset_index(drop=True)
for i, image_id in enumerate(df_meta.image_id):
    df_meta.loc[i, 'image_id'] = f'{image_id}_image'
df_meta.columns = ['id', 'dim0', 'dim1', 'split']
df_meta
```

```python
df = pd.merge(df, df_meta, on='id', how='left')
df
```

```python
save_path = './data-640/label_txt'

if not os.path.exists(save_path):
    os.makedirs(save_path)

for i in range(df.shape[0]):
    id = df.loc[i, 'id']
    with open(f'{save_path}/{id}.txt', 'w') as f:
        label_list = df.loc[i, 'label'].split()
        box_num = int(len(label_list)/6)
        if label_list[0] == 'opacity':
            dim0 = df.loc[i, 'dim0']
            dim1 = df.loc[i, 'dim1']
            for j in range(box_num):
                x1, y1, x2, y2 = [float(num) for num in label_list[6*j+2:6*j+6]]
                x_mid = str((x1 + x2) / 2 / dim1)
                y_mid = str((y1 + y2) / 2 / dim0)
                w = str((x2 - x1) / dim1)
                h = str((y2 - y1) / dim0)
                f.write(f'0 {x_mid} {y_mid} {w} {h} \n')
```
